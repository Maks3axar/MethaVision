```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from ultralytics import YOLO
import torch

model = YOLO("runs/segment/train/weights/best.pt")

image = cv2.imread("YOLO_dataset/test/images/AxioCamMRc512.jpg")

results = model(image, imgsz=640, iou=0.4, conf=0.4)

# detections = model.predict(image)





```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Предполагается, что 'image' и 'results' уже определены в вашем коде

classes = results[0].boxes.cls.cpu().numpy()
class_names = results[0].names

masks = results[0].masks.data
num_masks = masks.shape[0]

# Генерируем уникальные цвета для каждого класса
unique_classes = np.unique(classes)
colors = {cls: tuple(np.random.randint(0, 256, 3).tolist()) for cls in unique_classes}

class_counts = {}

labeled_image = image.copy()

for i in range(num_masks):
    cls = classes[i]
    color = colors[cls]
    mask = masks[i].cpu()
    
    mask_resized = cv2.resize(np.array(mask), (image.shape[1], image.shape[0]), interpolation=cv2.INTER_NEAREST)
    mask_contours, _ = cv2.findContours(mask_resized.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    if len(mask_contours) > 0:
        cv2.drawContours(labeled_image, mask_contours, -1, color, 2)
    
    # Увеличиваем счетчик для соответствующего класса
    if cls in class_counts:
        class_counts[cls] += 1
    else:
        class_counts[cls] = 1

# Подготовка к отображению списка классов и количества найденных объектов
start_x, start_y = 10, 30  # Начальная позиция для текста
font_scale = 1
font_thickness = 3

for idx, cls in enumerate(unique_classes):
    class_name = class_names[cls]
    count = class_counts.get(cls, 0)
    text = f"{class_name}: {count}"
    color = colors[cls]
    
    cv2.putText(labeled_image, text, (start_x, start_y + idx * 30), cv2.FONT_HERSHEY_SIMPLEX, font_scale, color, font_thickness)

# Отображение обработанного изображения
plt.figure(figsize=(8, 8), dpi=150)
labeled_image = cv2.cvtColor(labeled_image, cv2.COLOR_BGR2RGB)
plt.imshow(labeled_image)
plt.axis('off')
plt.show()
```
